```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import re
```

```python
df = pd.read_csv("../data/vacancies.csv")
# len(df)
df.head()
```

```python
def remove_russian_words(text):
    return re.sub(r'\b[а-яА-ЯёЁ]+\b', '', text)

df["text"] = df["description"].fillna('').apply(remove_russian_words)
```

```python
vectorizer = TfidfVectorizer(
    max_features=3000,
    token_pattern=r'\b[a-zA-Z]{2,}\b',  # только латинские слова длиной ≥2
    stop_words='english',
    min_df=3,
    max_df=0.9
)
X = vectorizer.fit_transform(df["text"])
```

```python
n_clusters = 8
model = KMeans(n_clusters=n_clusters, random_state=42)
df["cluster"] = model.fit_predict(X)
```

```python
cluster_labels = {
    0: "Data Science",
    1: "Backend",
    2: "Frontend",
    3: "DevOps",
    4: "GameDev"
}

# Пример сопоставления в легенде
import matplotlib.patches as mpatches
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X_reduced = PCA(n_components=2).fit_transform(X.toarray())

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=df["cluster"], cmap='tab10')

# Добавляем легенду вручную
handles = [mpatches.Patch(color=scatter.cmap(scatter.norm(i)), label=label)
           for i, label in cluster_labels.items()]
plt.legend(handles=handles, title="Кластеры")

plt.title("Кластеры вакансий")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
def clean_text(text):
    text = str(text).lower()                          # Приводим к нижнему регистру
    text = re.sub(r'\\[ntr]', ' ', text)              # Удаляем \n, \t и др.
    text = re.sub(r'[^a-zA-Zа-яА-Я0-9\s]', ' ', text)  # Удаляем пунктуацию
    text = re.sub(r'\s+', ' ', text)                  # Убираем лишние пробелы
    # text = re.sub(r'\b[а-яА-ЯёЁ]+\b', '', text)
    return text.strip()



# df["cluster"] = model.predict(X)

# Предсказание кластера для кандидата
candidate_stack = clean_text("json,qa")
print(candidate_stack)
candidate_vector = vectorizer.transform([candidate_stack])
predicted_cluster = model.predict(candidate_vector)[0]
print("->Предсказанный кластер:", predicted_cluster)
# Фильтруем вакансии из того же кластера
matching_vacancies = df[df["cluster"] == predicted_cluster]

# Показываем только названия вакансий
print(matching_vacancies[["title", "url"]].dropna().head(10).to_string(index=False))

```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, df["cluster"])
print(f"Силуэтный коэффициент: {score}")
```

```python
for i in range(n_clusters):
    cluster_center = model.cluster_centers_[i]
    top_words = [vectorizer.get_feature_names_out()[idx] for idx in cluster_center.argsort()[-10:]]
    print(f"Кластер {i}: {top_words}")
```

```python
inertia = model.inertia_
print(f"Инерция: {inertia}")
```

```python
inertias = []
for k in range(2, 16):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    inertias.append(model.inertia_)
# Визуализация
import matplotlib.pyplot as plt
plt.plot(range(2, 16), inertias, marker='o')
plt.xlabel('Число кластеров')
plt.ylabel('Инерция')
plt.show()
```

```python
silhouette_scores = []
for k in range(2, 16):
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
# Визуализация
plt.plot(range(2, 16), silhouette_scores, marker='o')
plt.xlabel('Число кластеров')
plt.ylabel('Силуэтный коэффициент')
plt.show()
```

```python
import matplotlib.pyplot as plt
cluster_counts = df["cluster"].value_counts().sort_index()
plt.bar(range(n_clusters), cluster_counts)
plt.xlabel('Кластер')
plt.ylabel('Количество вакансий')
plt.title('Распределение вакансий по кластерам')
plt.show()
```
